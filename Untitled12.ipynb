```python
m_x=np.mean(x)
m_y=np.mean(y)
```

```python
l=len(x)
sum1=0
sum2=0
for i in range(l):
    sum1+=(x[i]-m_x)*(y[i]-m_y)
    sum2+=(x[i]-m_x)**2
print(sum1,sum2)
m=sum1/sum2
c=m_y-m*(m_x)
print(m,c)

    
```

```python
max_x=np.max(x)+100
min_x=np.min(y)-100
X=np.linspace(min_x,max_x,100)
Y=c+m*X
#plt.scatter(x,y,color="red")
plt.plot(X,Y,color="blue",label="Regression Line")
plt.scatter(x,y,color="red")
plt.legend()
plt.show()
```
